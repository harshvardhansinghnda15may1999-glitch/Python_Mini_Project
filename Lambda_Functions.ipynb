```python
#Detect Palindromes in a list


def is_palindrome(word):
    return word == word[::-1]
words = ['level', 'python', 'radar', 'world', 'madam']
palindromes = [word for word in words if is_palindrome(word)]

print("Palindromes found:", palindromes)
```

```python
# Filter Out Negative Using lamba and filter()

def is_negative(num):
  if num >= 0:
    return True
  else:
    return False

nums = [-5,0,3,-2,8,-1]
negative = [num for num in nums if is_negative(num)]
print("Non Negative :",negative)
```

```python
# Extract words Starting with Vowels Using lamba and filter()

def is_vowel(word):
  if word[0].lower() in 'aeiou':
    return True
  else:
    return False

words = ['apple','banana','orange','grape','umbrella','cherry']
vowel_words = [word for word in words if is_vowel(word)]
print("Vowel Words :",vowel_words)
```

```python
#Tag Even and Odd Numbers Using lamba and map()

def tag_even_odd(num):
  if num % 2 ==0:
    return True
  else:
    return False

nums = [10,13,8,5,22]
numbers = []
for num in nums:
  if tag_even_odd(num) == True:
    numbers.append(f'{num} is even')
  else:
    numbers.append(f'{num} is odd')

print(numbers)
```

```python
# Filter Names That Have Vowels in All Positions Usings lambda and filter()


def only_vowel(word):
  for char in word:
    if char.lower() not in 'aeiou':
      return False
  return True

words = ['aeiou', 'idea', 'queue','sky', 'oui', 'ea']

print("Names with only vowels:", list(filter(only_vowel, words)))
```

```python
# Multiply All Even Numbers in Nested Lists Using lamba, map(), and filter()

def is_even(num):
  return num % 2 == 0

data = [[1,2,3],[4,5,6],[7,8]]
all_multiplied_evens = []

for sublist in data:
  even_numbers_in_sublist = list(filter(is_even, sublist))

  multiplied_numbers_in_sublist = list(map(lambda x: x * 3, even_numbers_in_sublist))

  all_multiplied_evens.extend(multiplied_numbers_in_sublist)

print("Multiplied even numbers:", all_multiplied_evens)
```

```python
# Identify Numbers Whose Digits Are All Even

nums = [242, 860,123,444,345]

def filter_out(num):
  if num % 2 == 0:
    return True

result = list(filter(filter_out,nums))
print(result)

#
```
