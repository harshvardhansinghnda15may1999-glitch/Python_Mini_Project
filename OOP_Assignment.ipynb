```python
# Student Record Management System

class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no
        self.marks = {}

    # ── Method 1 ──────────────────────────────────────────
    def add_subject(self, subject, score):
        self.marks[subject] = score
        print(f"  ✔ '{subject}' added with score {score}")

    # ── Method 2 ──────────────────────────────────────────
    def average(self):
        if not self.marks:
            return 0
        return sum(self.marks.values()) / len(self.marks)

    # ── Method 3 ──────────────────────────────────────────
    def display(self):
        print("=" * 35)
        print(f"  Name    : {self.name}")
        print(f"  Roll No : {self.roll_no}")
        print("-" * 35)
        print("  Subject          | Marks")
        print("-" * 35)
        for subject, score in self.marks.items():
            print(f"  {subject:<17}| {score}")
        print("-" * 35)
        print(f"  Average Marks   : {self.average():.2f}")
        print("=" * 35)


# ── Testing the class ──────────────────────────────────────
s1 = Student("Rahul Sharma", 101)

s1.add_subject("Math",    92)
s1.add_subject("Science", 88)
s1.add_subject("English", 75)
s1.add_subject("Hindi",   83)

s1.display()
```

```python
# Library Book Tracker

class LibraryBook:
    def __init__(self, title, author, available_copies):
        self.title             = title
        self.author            = author
        self.available_copies  = available_copies
        self.borrowed_by       = []

    # ── Method 1 ──────────────────────────────────────────
    def borrow_book(self, name):
        if self.available_copies > 0:
            self.available_copies -= 1
            self.borrowed_by.append(name)
            print(f"  '{name}' successfully borrowed '{self.title}'")
        else:
            print(f"  Sorry '{name}', No copies of '{self.title}' are available right now.")

    # ── Method 2 ──────────────────────────────────────────
    def return_book(self, name):
        if name in self.borrowed_by:
            self.available_copies += 1
            self.borrowed_by.remove(name)
            print(f"   '{name}' successfully returned '{self.title}'")
        else:
            print(f"   '{name}' has no record of borrowing '{self.title}'")

    # ── Method 3 ──────────────────────────────────────────
    def status(self):
        print("=" * 40)
        print(f"   Title            : {self.title}")
        print(f"    Author           : {self.author}")
        print(f"   Available Copies : {self.available_copies}")
        print("-" * 40)
        if self.borrowed_by:
            print("   Borrowed By:")
            for i, person in enumerate(self.borrowed_by, start=1):
                print(f"      {i}. {person}")
        else:
            print("   Borrowed By      : None")
        print("=" * 40)


# ── Testing the class ──────────────────────────────────────
book = LibraryBook("The Alchemist", "Paulo Coelho", 3)

print("\n--- Initial Status ---")
book.status()

print("\n--- Borrowing Books ---")
book.borrow_book("Rahul")
book.borrow_book("Priya")
book.borrow_book("Aman")
book.borrow_book("Sneha")       # Should show unavailable

print("\n--- Status After Borrowing ---")
book.status()

print("\n--- Returning a Book ---")
book.return_book("Priya")
book.return_book("Ravi")        # Never borrowed — error case

print("\n--- Final Status ---")
book.status()
```

```python
# Volume Calculator for 3D Shapes

import math

class VolumeCalculator:

    def calculate(self, shape, *args):

        shape = shape.lower().strip()

        # ── Cube ──────────────────────────────────────────
        if shape == "cube":
            a = args[0]
            volume = a ** 3
            print(f"   Cube")
            print(f"     Side          : {a}")
            print(f"     Formula       : a³")
            print(f"     Volume        : {a}³ = {volume:.4f}")

        # ── Cuboid ────────────────────────────────────────
        elif shape == "cuboid":
            l, w, h = args[0], args[1], args[2]
            volume = l * w * h
            print(f"   Cuboid")
            print(f"     Length        : {l}")
            print(f"     Width         : {w}")
            print(f"     Height        : {h}")
            print(f"     Formula       : l × w × h")
            print(f"     Volume        : {l} × {w} × {h} = {volume:.4f}")

        # ── Cylinder ──────────────────────────────────────
        elif shape == "cylinder":
            r, h = args[0], args[1]
            volume = math.pi * r ** 2 * h
            print(f"   Cylinder")
            print(f"     Radius        : {r}")
            print(f"     Height        : {h}")
            print(f"     Formula       : π × r² × h")
            print(f"     Volume        : π × {r}² × {h} = {volume:.4f}")

        # ── Sphere ────────────────────────────────────────
        elif shape == "sphere":
            r = args[0]
            volume = (4/3) * math.pi * r ** 3
            print(f"   Sphere")
            print(f"     Radius        : {r}")
            print(f"     Formula       : (4/3) × π × r³")
            print(f"     Volume        : (4/3) × π × {r}³ = {volume:.4f}")

        # ── Unknown Shape ─────────────────────────────────
        else:
            print(f"   Unknown shape '{shape}'.")
            print(f"     Supported: cube, cuboid, cylinder, sphere")
            return None

        print(f"  {'='*38}")
        print(f"    Final Volume   : {volume:.4f} cubic units")
        print(f"  {'='*38}\n")
        return round(volume, 4)


# ── Testing ───────────────────────────────────────────────
calc = VolumeCalculator()

print("=" * 42)
print("       VOLUME CALCULATOR - TEST CASES      ")
print("=" * 42, "\n")

calc.calculate("cube",     5)
calc.calculate("cuboid",   4, 6, 3)
calc.calculate("cylinder", 7, 10)
calc.calculate("sphere",   9)
calc.calculate("pyramid")           # unknown shape test
```

```python
# User Authentication System with Menu Options

class UserDirectory:
    def __init__(self):
        self.users = {}          # { username: password }

    # ── Create Account ────────────────────────────────────
    def create_account(self):
        print("\n  ── Create Account ──────────────────")
        username         = input("  Enter Username         : ").strip()
        password         = input("  Enter Password         : ").strip()
        confirm_password = input("  Confirm Password       : ").strip()

        # Check 1: Username already exists
        if username in self.users:
            print(f"\n   Username '{username}' is already registered.")

        # Check 2: Passwords don't match
        elif password != confirm_password:
            print("\n   Passwords do not match. Please try again.")

        # All checks passed → Register user
        else:
            self.users[username] = password
            print(f"\n   Account created successfully! Welcome, {username}!")

    # ── Sign In ───────────────────────────────────────────
    def sign_in(self):
        print("\n  ── Sign In ─────────────────────────")
        username = input("  Enter Username         : ").strip()
        password = input("  Enter Password         : ").strip()

        # Check credentials
        if username in self.users and self.users[username] == password:
            print(f"\n   Successfully Logged In! Welcome back, {username}!")
        else:
            print("\n   Incorrect username or password.")

    # ── Main Menu ─────────────────────────────────────────
    def run(self):
        while True:
            print("\n" + "=" * 40)
            print("        USER DIRECTORY SYSTEM         ")
            print("=" * 40)
            print("  1. Create Account")
            print("  2. Sign In")
            print("  3. Exit")
            print("-" * 40)

            choice = input("  Choose an option (1/2/3)  : ").strip()

            if choice == "1":
                self.create_account()

            elif choice == "2":
                self.sign_in()

            elif choice == "3":
                print("\n   Goodbye! Exiting the system.\n")
                break

            else:
                print("\n   Invalid option. Please enter 1, 2, or 3.")


# ── Run the program ───────────────────────────────────────
directory = UserDirectory()
directory.run()
```
