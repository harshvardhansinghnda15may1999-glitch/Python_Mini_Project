```python
# Tuple Time Travel

Years = (2001,2010,1999,2023)
print(Years[0], Years[(len(Years)-1)])


#Tuple of Tuples

locations = ((0,0),(1,2),(3,5),(4,4))
print(locations[1])
print(locations[len(locations)-1][0],locations[len(locations)-1][1])
```

```python
# Sets - No Duplicates Allowed!

# Unique or Not?

numbers = [2,4,4,6,2,8]
print("Tuple -: ", numbers)
unique_numbers = set(numbers)
print("Set -: ", unique_numbers)


#Set Magic Tricks

a= {1,2,3,4}
b= {3,4,5,6}
print("Union :- ",a|b)
print("Intersection :- ",a&b)
print("Difference",a-b)
print("Not unique :- ",a^b)
```

```python
# Dictionaries - Look It Up Like a Pro

# Superhero Database

heroes = {"Spider-Man": "Peter parker", "Batman": "Bruce Wayne"}
print(heroes["Spider-Man"])

heroes["Wonder Woman"] = "Diana Prince"
print(heroes["Wonder Woman"])
print(heroes)

heroes["Batman"] = "Bruce Bayne"
print(heroes)
del heroes["Batman"]
print(heroes)

# Dictionary Challenge - Student Grades

students = {"Aditya": 85, "Anshu": 92, "Yash": 78}
print("Students Names :", students.keys())
print("Students Grades :", students.values())
print(students.items())
```

```python
# Voting App Lite

votes = ["Harsh","Harsh","Raj","Raj","Raj"]

vote_counts = {}
for vote in votes:
    vote_counts[vote] = vote_counts.get(vote, 0) + 1

print("Vote Counts:", vote_counts)
```
