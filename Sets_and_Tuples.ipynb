```python
# Tuple Indexing and Slicing

t= (5,10,15,20,25)
print(t[-3:])

# Check Duplicates Using Sets

duplicates = []
seen = set()

for item in t:
    if item in seen:
        print("Yes")
    else:
        seen.add(item)

# Tuple Immutability Test

# t[2] = 30  # This line causes a TypeError because tuples are immutable.

# Remove Specific Element from Set

my_set = {1, 2, 3, 4, 5}
element_to_remove = 4
if element_to_remove in my_set:
    my_set.remove(element_to_remove)

print(my_set)

# Find Union of Two Sets

set1 ={1,2}
set2 = {2,3}
union_set = set1.union(set2)
print(union_set)

#Swap Variables Using Tuple Unpacking

a = (5,6,7,8,9)
b = (10,11,12,13,14)
a, b = b, a
print(a, b)

#Set Difference

A = {1,2,3}
B = {3,4,5}
difference_set = A.difference(B)
print(difference_set)

# Convert Tuple to List and Modify (Replace element)

my_tuple = (1, 2, 3)
my_list = list(my_tuple)
index = my_list.index(2)
my_list[index] = 20
my_tuple = tuple(my_list)
print(my_tuple)

# Get Common Elements of Two Sets

set1 = {1,2,3}
set2 = {2,4}
common_elements = set1.intersection(set2)
print(common_elements)

# TUPLE OF TUPLES - ACCESS ELEMENT

my_tuple = ((5, 6), (7,8,9))
number = 9
if any(number in inner_tuple for inner_tuple in my_tuple):
  print("Element found in the tuple.")
else:
  print("Element not found in the tuple.")

# Loop Through Set and Print Squares

sets = {1,2,3}
for num in sets:
  print(num ** 2)

# Sets form Tuple

t = (1,2,2,3)
s = set(t)
print(s)

# Tuple Concatenation

t1 = (1,2)
t2 = (3,4)
t3 = t1 + t2
print(t3)

#Compare two sets for Equality

s1 = {1,2,3}
s2 = {3,2,1}
if s1 == s2:
  print("Equal sets")

# Get Unique Tuples from List

my_list = [(1, 2), (2, 3), (1, 2)]
unique_tuples = set(my_list)
print(unique_tuples)
```
