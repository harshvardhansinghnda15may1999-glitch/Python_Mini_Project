```python
# Group Anagrams

def groupAnagrams(strs):
    anagram_map = {}
    for word in strs:
        sorted_word = "".join(sorted(word))
        if sorted_word not in anagram_map:
            anagram_map[sorted_word] = []
        anagram_map[sorted_word].append(word)
    return list(anagram_map.values())

print(groupAnagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
print(groupAnagrams(["abc", "bca", "cab", "xyz", "zyx"]))
```

```python
# Top K Frequent Elements

import collections

def topKFrequent(nums, k):
    # Count the frequency of each number
    counts = collections.Counter(nums)

    # Group numbers by their frequency
    freq_map = [[] for _ in range(len(nums) + 1)]
    for num, freq in counts.items():
        freq_map[freq].append(num)

    # Collect the k most frequent elements
    result = []
    for i in range(len(freq_map) - 1, 0, -1):
        for num in freq_map[i]:
            result.append(num)
            if len(result) == k:
                return result
    return result

print(topKFrequent([1,1,1,2,2,3], 2))
print(topKFrequent([4,4,5,5,5,6], 1))
```

```python
#  Isomorphic Strings

def isIsomorphic(s, t):
    if len(s) != len(t):
        return False

    s_map = {}
    t_map = {}

    for char_s, char_t in zip(s, t):
        if char_s not in s_map and char_t not in t_map:
            s_map[char_s] = char_t
            t_map[char_t] = char_s
        elif s_map.get(char_s) != char_t or t_map.get(char_t) != char_s:
            return False

    return True

print(isIsomorphic("egg", "add"))
print(isIsomorphic("foo", "bar"))
```

```python
# Word Pattern

def wordPattern(pattern, s):
    words = s.split()
    if len(pattern) != len(words):
        return False

    char_to_word = {}
    word_to_char = {}

    for char_p, word_s in zip(pattern, words):
        if char_p not in char_to_word and word_s not in word_to_char:
            char_to_word[char_p] = word_s
            word_to_char[word_s] = char_p
        elif char_to_word.get(char_p) != word_s or word_to_char.get(word_s) != char_p:
            return False
    return True

print(wordPattern("abba", "dog cat cat dog"))
print(wordPattern("abba", "dog cat cat fish"))
```

```python
# Word Frequency in Paragraph

import collections
import re

def wordFrequency(text):
    # Convert to lowercase and remove punctuation
    text = text.lower()
    text = re.sub(r'[\W_]+', ' ', text) # Replace non-alphanumeric with space

    # Split into words and count frequencies
    words = text.split()
    return collections.Counter(words)

print(wordFrequency("Hello hello world!"))
print(wordFrequency("Python is fun. Python is powerful."))
```

```python
import collections

def uniqueOccurrences(arr):
    # Count the frequency of each number
    counts = collections.Counter(arr)

    # Check if all frequencies are unique
    return len(counts.values()) == len(set(counts.values()))

print(uniqueOccurrences([1,2,2,1,1,3]))
print(uniqueOccurrences([1,2,3]))
```

```python
# Most Common Non-Banned Word

import collections
import re

def mostCommonWord(paragraph, banned):
    # Normalize the paragraph: lowercase and remove punctuation
    # Using a robust regex to remove any character that is not alphanumeric or whitespace
    normalized_paragraph = re.sub(r'[^a-zA-Z0-9\s]', ' ', paragraph.lower())
    words = normalized_paragraph.split()

    # Count word frequencies, excluding banned words
    word_counts = collections.Counter(word for word in words if word not in banned)

    # Find the most common word
    # Handle case where word_counts might be empty after banning
    if not word_counts:
        return ""

    return word_counts.most_common(1)[0][0]


print(mostCommonWord("Bob hit a ball, the hit BALL flew far.", ["hit"]))
# 'ball'
print(mostCommonWord("Jack and Jill went up the hill. Jill fell.", ["jill"]))
```

```python
# First Unique Character in a String

import collections

def firstUniqChar(s):
    # Build character frequency map
    char_counts = collections.Counter(s)

    # Iterate through the string to find the first character with a count of 1
    for i, char in enumerate(s):
        if char_counts[char] == 1:
            return i

    return -1

print(firstUniqChar("leetcode")) # 0
print(firstUniqChar("aabbcc")) # -1
```
