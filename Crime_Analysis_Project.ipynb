```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.columns
```

```python
df.columns = [
    "Serial_No",
    "State",
    "Crime_2020",
    "Crime_2021",
    "Crime_2022",
    "Population_2022",
    "Crime_Rate_2022",
    "Chargesheet_Rate_2022"
]
top_states = df.sort_values(by="Crime_2022", ascending=False)
top10 = top_states.head(10)

top10
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(top10["State"], top10["Crime_2022"])
plt.xticks(rotation=45)
plt.title("Top 10 States with Highest Crime (2022)")
plt.xlabel("State")
plt.ylabel("Crime Count")
plt.show()
```

```python
# The analysis indicates that Uttar Pradesh recorded the highest number of crimes in 2022 among all states. This trend can be attributed to factors such as its large population, rapid urbanization, and possibly more comprehensive crime reporting mechanisms. The findings highlight the need for targeted policing strategies, improved law enforcement resource allocation, and preventive measures in high-crime regions to enhance public safety.
```

```python
top10[["State", "Crime_2022"]]
```

```python
total_2020 = df["Crime_2020"].sum()
total_2021 = df["Crime_2021"].sum()
total_2022 = df["Crime_2022"].sum()

print(total_2020, total_2021, total_2022)
```

```python
years = ["2020", "2021", "2022"]
crime_values = [total_2020, total_2021, total_2022]

import matplotlib.pyplot as plt

plt.figure()
plt.plot(years, crime_values, marker='o')
plt.title("Crime Trend in India (2020–2022)")
plt.xlabel("Year")
plt.ylabel("Total Crime")
plt.show()
```

```python
# A steady decline in crime is observed between 2020 and 2022, with total cases dropping significantly. While this may indicate improvements in law enforcement and public safety, it is also likely influenced by pandemic-related restrictions that limited social and economic activity. Therefore, further analysis is required to determine whether this trend reflects genuine improvement or temporary external factors.
```

```python
plt.figure()

plt.scatter(df["Population_2022"], df["Crime_2022"])

plt.title("Crime vs Population (2022)")
plt.xlabel("Population (in Lakhs)")
plt.ylabel("Crime Count")

plt.show()
```

```python
# Although larger populations tend to be associated with higher crime counts, the relationship is not strictly proportional. The presence of outliers—states with lower populations but comparatively high crime—highlights that crime is influenced by multiple structural factors beyond population size. This emphasizes the need for targeted, data-driven policing strategies rather than relying solely on population-based resource allocation.
```

```python
plt.figure()

plt.scatter(df["Population_2022"], df["Crime_2022"])

# Only top 5 states
top_states = df.sort_values(by="Crime_2022", ascending=False).head(5)

for i in top_states.index:
    plt.text(df["Population_2022"][i], df["Crime_2022"][i], df["State"][i], fontsize=8)

plt.xlabel("Population")
plt.ylabel("Crime")
plt.title("Crime vs Population (2022)")

plt.show()
```

```python
# Although population size influences total crime, the scatter plot reveals noticeable deviations from a linear trend. For instance, Delhi demonstrates high crime with comparatively lower population, highlighting the impact of urban concentration and reporting intensity. This suggests that policy decisions should focus on region-specific factors rather than relying solely on population-based resource allocation.
```
