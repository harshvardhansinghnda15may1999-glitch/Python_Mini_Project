```python
#Strings Playing with Text

#Name Spellcraft

name = input('Enter your name')
uppercase = name.upper()
print(uppercase)
lowercase = name.lower()
print(lowercase)
Titlecase = name.title()
print(Titlecase)
words = name.split()
word_count = len(words)
print(f"The number of words in '{name}' (excluding spaces) is: {word_count}")

#Code your Mood

sentence = "I love Python programming!"
replace = sentence.replace("Python", "Java")
print(replace)
find = sentence.find("love")
if (find == -1):
    print("Not Found")
else:
    print("Found  at", find)


# Slice of the story

text = "Learning Python is fun!"
print(text[0:8])
print(text[-4:])
print(text[::-1])
```

```python
# Bonus Challenge (Optional but Epic!)

#Secret String Puzzle

puzzle = input("Enter the string : ")
lowercase_puzzle = puzzle.lower()
print(lowercase_puzzle)

no_spaces_puzzle = puzzle.replace(" ", "")
print(f"String with no spaces: {no_spaces_puzzle}")

vowels = "aeiou"
vowel_count = 0
for char in lowercase_puzzle:
    if char in vowels:
        vowel_count += 1
print(f"Number of vowels: {vowel_count}")
```
